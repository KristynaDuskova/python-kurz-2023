```python
import pandas
import matplotlib.pyplot as plt
```

```python
graf = pandas.read_csv("platy_2021_02.csv")
graf
```

```python
graf["plat"].plot(
    kind="bar",
    color="green",
    width=0.5,
    ylabel= "plat",
    xlabel="uco",
    title="Plat zaměstnanců",
)

plt.show()
```

Když se snažím dát hist (viz.tabulka výše), tak mi to nefunguje. Nevím proč. 

```python
color = "green"
graf.plot.hist(bins=[30000, 35000, 40000, 45000, 50000, 55000, 60000])

plt.show()
```

```python
color = "green"

fig, ax = plt.subplots(figsize=(15, 7))
print(type(fig))
print(type(ax))


ax.set_title("Výplata zaměstnanců", fontweight="bold", fontsize=24)

ax.tick_params(axis="x", labelrotation=90)
ax.set_ylabel("Plat", color="darkgrey", fontsize=10)
ax.bar(graf.index, graf.plat, color=color)
plt.axhline(y=30000, color="k", linestyle="--")

for i in range(len(graf.plat)):
    plt.annotate(
        str(graf.plat[i]), xy=(graf.index[i], graf.plat[i]), ha="center", va="bottom"
    )

ax.spines["right"].set_visible(False)
```
